```python
import chant
import pandas as pd

def display_percent(x):    # used for easier-to-read probability tables
    if x == 0:
        return ''
    else:
        return str(int(x*100)) + '%'
 
nd = chant.noteData()
cd = chant.chantData()
```

```python
notes = chant.noteData()[['chantID', 'rint', 'reg_rel', 'pc_rel']] # take only columns you need!
notes['pitch'] = notes['reg_rel'] + '.' + notes['pc_rel'] 
notes.drop(['reg_rel','pc_rel'], axis=1) # drop columns you're done with!


for m in [chant.basicAuthentics, chant.basicPlagals]:
    with pd.option_context('display.float_format', display_percent, 'display.max_rows', 20):

        # generate a dataframe with just the rows from chantData that we'll need

        selected = cd[cd.modus.isin(m)][['chantID','modus']]   # alternate syntax to the .query() method used above

        # use pandas merge() to make a new dataframe combining the 'selected' and 'results' dataframes

        results = selected.merge(notes, on ='chantID')

        # make the crosstab    

    #     display(pd.crosstab(results.pitch, results.rint, normalize='index').sort_index(ascending=False))
    
        display(pd.crosstab(results.pitch, results.modus, normalize = 'columns').sort_index(ascending=False))
    
del(notes, results, selected)
```
